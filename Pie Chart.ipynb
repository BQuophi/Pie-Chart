```python
# Program that illustrates the construction a pie chart

import matplotlib.pyplot as plt
sections = 'Geography', 'Acturial Science', 'Data Science', 'Mathematics'
sizes = [180, 250, 300, 200]
fig1, axl = plt.subplots()
axl.pie(sizes, labels = sections, autopct = '%1.1f%%', shadow = True, startangle = 90)
axl.axis('equal')
plt.show()
```
